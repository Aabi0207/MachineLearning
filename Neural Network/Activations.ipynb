## Lets plot some of the activations

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
X = np.linspace(-5, 5, 100)
```

## plot Activations

```python
import tensorflow as tf
```

```python
g_linear = tf.keras.activations.linear(X)
g_sigmoid = tf.keras.activations.sigmoid(X)
g_relu = tf.keras.activations.relu(X)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

ax[0].plot(X, g_linear, lw=3, c="b")
ax[0].set_title("Linear Activation")
ax[0].axvline(x=0)

ax[1].plot(X, g_sigmoid, lw=3, c="b")
ax[1].set_title("Sigmoid Activation")
ax[1].axvline(x=0)

ax[2].plot(X, g_relu, lw=3, c="b")
ax[2].set_title("ReLU Activation")
ax[2].axvline(x=0)
```
