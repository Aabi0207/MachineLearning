### Problem Statement
     form a vector with the size of the house in 1000's SQ. FT. and form an another vector having house prices stored according to the sizes specified in the size vector using the numpy np.array function

```python
x_train = np.array([1, 2])  # The x_train variable usually contains Features of the traning data
y_train = np.array([300, 500]) # y_train variable usually contains Target value for the predictions

print(f"The sizes of the homes in the training datasets are {1000 * x_train}")
print(f"The prices of the homes in the training datasets are {1000 * y_train}")
```

## Plotting the data using Scatter plot from the matplotlib library

```python
# plt.figure(figsize=(4,8), dpi=200)
plt.scatter(x_train, y_train, marker="X", c="crimson")
plt.title("Size vs Price of the houses")
plt.ylabel("Prices of Houses in 1000's of â‚¹")
plt.xlabel("Size of Houses in 1000's of square feet.")
plt.show()
```

### Determine the values of the parameters w, and b such that the function predicts the most appropriate values

```python
w = 200
b = 100
```

### Write a function compute_model_output() which determines the values of regression model

```python
m = x_train.shape
temp = np.zeros(m)

def compute_model_output(w, b, x_train):
    for i in range(m[0]):
        f_wb = w * x_train[i] + b
        temp[i] = f_wb
    print(temp)
    return temp
```

```python
temp_values = compute_model_output(w, b, x_train)
```

```python
plt.plot(x_train, temp_values, label="Model Prediction", c="b", linewidth=3)
plt.scatter(x_train, y_train, label="Actual values", c="r", marker="X")

plt.title("Size of Houses vs the Prices of the Houses")
plt.xlabel("Size of House in 1000's sq.ft.")
plt.ylabel("Price of House in 1000's of â‚¹")

plt.show()
```
