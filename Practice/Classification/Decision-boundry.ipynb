# Logistic Regression (Decision Boundry)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

## Dataset

Let's suppose you have following training dataset
- The input variable `X` is a numpy array which has 6 training examples, each with two features
- The output variable `y` is also a numpy array with 6 examples, and `y` is either `0` or `1`

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

## Plot data

```python
fig, ax = plt.subplots(1,1)

plt.xlabel("$X_0$")
plt.ylabel("$X_1$")

pos = y == 1
neg = y == 0
pos = pos.reshape(-1,)  #work with 1D or 1D y vectors
neg = neg.reshape(-1,)

# Plot examples
ax.scatter(X[pos, 0], X[pos, 1], marker='x', s=80, c = 'red', label="1")
ax.scatter(X[neg, 0], X[neg, 1], marker='o', s=100, label="0", facecolors='none', edgecolors="blue", lw=3)
ax.legend()

ax.figure.canvas.toolbar_visible = False
ax.figure.canvas.header_visible = False
ax.figure.canvas.footer_visible = False

plt.show()
```

## Logistic Regression model

The Logistic Regression model predicts the values using the formula
$$\begin{align*}f_{w, b}(X) = g(z) \end{align*}$$

where
$$\begin{align} g(z) = \frac{1}{1 + e^{-z}}\end{align}$$

where 
$$ z = w_0x_0 + w_1x_1 + b$$

The model predicts the values between y = 0 or y = 1  
Therefore **f_wb > 0.5 = 1**  
**f_wb <= 0.5 = 0**

## Plot the sigmoid function

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
z = np.arange(-10, 11)

fig, ax = plt.subplots(1, 1, figsize=(8, 3), sharey=True)
ax.plot(z, sigmoid(z), c="b", lw=3)
ax.set_title("Sigmoid Function")
ax.set_xlabel("Z")
ax.set_ylabel("Sigmoid Z")
plt.show()
```

## Plot the Decision Boundry

```python
def plot_data(X, y, ax, pos_label="y=1", neg_label="y=0", s=80, loc='best' ):
    """ plots logistic data with two axis """
    # Find Indices of Positive and Negative Examples
    pos = y == 1
    neg = y == 0
    pos = pos.reshape(-1,)  #work with 1D or 1D y vectors
    neg = neg.reshape(-1,)

    # Plot examples
    ax.scatter(X[pos, 0], X[pos, 1], marker='x', s=s, c = 'red', label=pos_label)
    ax.scatter(X[neg, 0], X[neg, 1], marker='o', s=s, label=neg_label, facecolors='none', edgecolors="blue", lw=3)
    ax.legend(loc=loc)

    ax.figure.canvas.toolbar_visible = False
    ax.figure.canvas.header_visible = False
    ax.figure.canvas.footer_visible = False
```

```python
x0 = np.arange(0, 6)
x1 = 3 - x0

fig, ax = plt.subplots(1, 1, figsize=(8, 3))
ax.plot(x0, x1, lw = 3)
ax.axis([0, 4, 0, 3.5])
ax.fill_between(x0, x1, alpha=0)
plot_data(X, y, ax)
ax.set_xlabel("$X_0$")
ax.set_ylabel("$X_1$")
plt.show()
```
