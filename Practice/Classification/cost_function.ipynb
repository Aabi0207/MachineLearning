## The cost function for the required Logistic Regression is given below

$$J(w, b) = \frac{1}{m}\sum\limits_{i = 0}^{n - 1}(L(f_{w, b}(x^{(i)}), y^{(i)}) $$

where $$L(f_{w, b}(x^{(i)}), y^{(i)}) = -y^{(i)}(log(f_{w, b}(x{(i)}) - (1 - y{(i)})(log(1 - f_{w, b}(x{(i)})$$

where $$f_{w, b}(x{(i)} = g(z)$$
$$z = \bar w \bar x^{(i)} + \bar b$$
$$g(z) = \frac{1}{1 + e^{(-\bar w \bar x^{(i)} + \bar b)}}$$

### Write a function to compute the logistic cost the function

```python
def compute_cost_logistic(X, y, w, b):
    m, n = X.shape
    g = sigmoid(X @ w + b)
    cost =np.sum((-y * np.log(g)) - ((1 - y) * np.log(1 - g))) / m
    return cost
```

```python
w_tmp = np.array([1,1])
b_tmp = -3
compute_cost_logistic(X_train, y_train, w_tmp, -3)
```
