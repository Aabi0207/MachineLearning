# Classification

### Import all the required imports

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

## Load the two data sets

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
x_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
pos = y_train == 1
neg = y_train == 0

pos = y_train2 == 1
neg = y_train2 == 0

fig, ax = plt.subplots(1, 2, figsize=(8, 3), sharey=True)

ax[0].set_ylim(-0.08,1.1)
ax[0].scatter(x_train[pos], y_train[pos], s=80, c="r", label="1", marker="X")
ax[0].scatter(x_train[neg], y_train[neg], s=80, c="b", label="0", marker="o")
ax[0].set_ylabel("Y")
ax[0].set_xlabel("X")
ax[0].legend()

ax[1].axis([0, 4, 0, 4])
ax[1].scatter(x_train2[:, 0][pos], x_train2[:, 1][pos], c="r", s = 80, marker="X")
ax[1].scatter(x_train2[:, 0][neg], x_train2[:, 1][neg], s = 80, marker="o")
ax[1].set_ylabel("X1")
ax[0].set_xlabel("X0")
plt.tight_layout()

plt.show()
```
